```python
#Step 03 - Input/Output Split
#importing pandas library for handling the tables (.csv)
import pandas as pd
```

```python
#Step 03 - Input/Output Split
#Reading the data from csv file and storing the values in the variable 'SalaryData'
SalaryData=pd.read_csv("Salary_Data.csv")
#SalaryData
```

```python
#Step 03 - Input/Output Split
#Assign the data from column 'YearsExperience' from the csv file to the Variable 'independent'

independent=SalaryData[["YearsExperience"]]
#independent
```

```python
#Step 03 - Input/Output Split
#Assign the data from column 'Salary' from the csv file to the Variable 'dependent'

dependent=SalaryData[["Salary"]]
#dependent
```

```python
#Step 04 - Split train and test dataset
#sklearn is the library for processing the train and test datasets

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.3,random_state=0)
```

```python
#Step 05 - Model Creation using training set
#LinearRegression function imported from the sklearn library and pass the training parameters
#'fit' instruction plays the importent role as it creats the brain

from sklearn.linear_model import LinearRegression
Regressor=LinearRegression()
Regressor.fit(x_train,y_train)
```

```python
#Checking the values of the 'Weight' and 'Bias'

print("Weight is", Regressor.coef_)
print("Bias is",Regressor.intercept_)
```

```python
#Step 06 - Model Eveluation using Test set
#Predicting the value. Provide the input and model will generate the predicted output
y_pred=Regressor.predict(x_test)
```

```python
#Step 06 - Model Evaluation using Test set
#Importing the 'metrics' for 'r2_score' function
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#Step 06 - Model Evaluation using Test set
#Result close to 1 is best Model; Result close to 0 is bad model

r_score
```

```python
#Step 07 - Save the best model
#pickle library used for saving the model. Create the file with a name
import pickle
FileName="finalized_model_linear.sav"
```

```python
#Step 07 - Save the best model
#Sending the mode to the created file. 'wb' - Write Binary
pickle.dump(Regressor,open(FileName,'wb'))
```

```python
#Step 08 - Load the saved model and test.
#Only the following part will be forwarded to next department for deployment
#Pass the input and model will predict the output for you.

loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb'))
result=loaded_model.predict([[13]])
```
